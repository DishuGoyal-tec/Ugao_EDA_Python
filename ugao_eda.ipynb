<h2 style="margin-top:0;color:#1f3fbf;">Project Overview</h2>

This project analyzes transactional sales data from <b>Ugaoo</b>, an Indian gardening and plant e-commerce brand.
The analysis focuses on understanding <b>sales performance, customer purchasing behavior, product category performance, regional demand patterns, and the impact of payment methods and discount strategies on profitability</b>.

By exploring historical order data, the objective is to uncover <b>key business insights</b> that can support better decision-making in pricing strategy, regional expansion, and product positioning.

<div style="background:#f4fbf6;border-left:6px solid #2e9d55;padding:24px;border-radius:8px;color:#000;font-family:Arial, Helvetica, sans-serif;line-height:1.6">

<h2 style="margin-top:0;color:#1e7a43;">Dataset Overview</h2>

The dataset contains transactional sales records from Ugaoo’s e-commerce platform covering the period <b>2023–2025</b>.
Each row represents an individual customer order including product, payment, and profitability information.

<b>Dataset Highlights</b>

• Total transactions analyzed: <b>18,000+ orders</b>
• Time period covered: <b>2023 – 2025</b>
• Granularity: <b>Order-level transaction data</b>

<b>Key Variables in the Dataset</b>

• Order Value
• Product Category
• State / Region
• Payment Method
• Discount Percentage
• Profit Generated
• Return Status

This dataset enables analysis of <b>profit drivers, customer behavior patterns, product performance, and operational risks</b> within Ugaoo’s online sales operations.

</div>


<div style="background:#fff7f2;border-left:6px solid #ff7a45;padding:22px;border-radius:8px;color:#000;line-height:1.6">

<h2 style="margin-top:0;color:#d9480f;">Business Problem</h2>

Ugaoo’s e-commerce business has been experiencing uneven performance across different states, product categories, and order types between <b>2023–2025</b>. While order volume has grown, the company lacks clear visibility into the factors driving <b>profitability, product returns, and regional sales performance</b>.

To address this challenge, the analysis focuses on answering the following key business questions:

• Which <b>states and regions generate the highest revenue and profit</b>
• Which <b>product categories drive the most business value</b>
• How <b>payment methods influence order returns</b>
• Whether <b>discount strategies are impacting profit margins</b>

This analysis aims to uncover these insights and support <b>data-driven decisions for pricing strategy, sales optimization, and operational efficiency</b>.

</div>

```python
# city wise orders
city_counts = df["city"].value_counts().head(10)

plt.figure(figsize=(10,5))

# Ugaoo green gradient palette for 10 bars
palette = [
"#1b5e20", "#2e7d32", "#388e3c", "#43a047", "#4caf50",
"#66bb6a", "#81c784", "#a5d6a7", "#c8e6c9", "#e8f5e9"
]

ax = sns.barplot(
    x=city_counts.index,
    y=city_counts.values,
    palette=palette
)

# value labels
for bar in ax.patches:
    ax.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        int(bar.get_height()),
        ha="center",
        va="bottom",
        fontsize=9,
        fontweight="bold"
    )

plt.xticks(rotation=45, ha="right")
plt.xlabel("City", fontsize=11)
plt.ylabel("Number of Orders", fontsize=11)
plt.title("Top 10 Cities by Number of Orders", fontsize=13, fontweight="bold")

# subtle grid
plt.grid(axis="y", linestyle="--", alpha=0.3)

# light natural background
plt.gca().set_facecolor("#f6fbf6")

plt.tight_layout()
plt.show()
```
